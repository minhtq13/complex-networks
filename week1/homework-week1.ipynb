```python
import networkx as nx
import matplotlib.pyplot as plt
# Load the Florentine Families graph
florentine = nx.florentine_families_graph()

# Visualize the graph
plt.figure(figsize=(8, 6)) # Adjust figure size for better viewing
nx.draw(florentine, with_labels=True, node_color='gold', node_size=800)
plt.title("Florentine Families Graph")
plt.show()

# Print some basic information about the graph
print("Number of nodes:", florentine.number_of_nodes())
print("Number of edges:", florentine.number_of_edges())
```

```python

deg_cent = nx.degree_centrality(florentine)
close_cent = nx.closeness_centrality(florentine)
bet_cent = nx.betweenness_centrality(florentine)
eig_cent = nx.eigenvector_centrality(florentine)
def top_n_ranking(metric, n=5):
  # Sort items by value in descending order and take the top N
  sorted_nodes = sorted(metric.items(), key=lambda x: x[1], reverse=True)[:n]
  # Extract just the node list (excluding the value for cleaner comparison)
  return [node for node, value in sorted_nodes]

# Get top 5 nodes for each centrality measure
top5_degree = top_n_ranking(deg_cent, n=5)
top5_closeness = top_n_ranking(close_cent, n=5)
top5_betweenness = top_n_ranking(bet_cent, n=5)
top5_eigenvector = top_n_ranking(eig_cent, n=5)

# Print and compare the rankings
print("Top 5 by Degree Centrality:", top5_degree)
print("Top 5 by Closeness Centrality:", top5_closeness)
print("Top 5 by Betweenness Centrality:", top5_betweenness)
print("Top 5 by Eigenvector Centrality:", top5_eigenvector)
```

```python
# Try directed or weighted versions
import networkx as nx
import matplotlib.pyplot as plt

# Ti  th gc khng c hng, khng c trng s
florentine_undirected = nx.florentine_families_graph()

# To mt  th mi s c trng s (v c th c hng nu cn)
# Hy to mt phin bn c trng s  minh ha
florentine_weighted = nx.Graph() # S dng nx.DiGraph() cho  th c hng

# Thm cc nt t  th gc
florentine_weighted.add_nodes_from(florentine_undirected.nodes())

# Thm cc cnh vi trng s ty   minh ha
# Trong mt kch bn thc t, bn s ti trng s t mt tp d liu
weighted_edges = [
    ('Acciaiuoli', 'Medici', 2),
    ('Castellani', 'Peruzzi', 3),
    ('Castellani', 'Strozzi', 1),
    ('Castellani', 'Barbadori', 2),
    ('Medici', 'Barbadori', 4),
    ('Medici', 'Ridolfi', 2),
    ('Medici', 'Tornabuoni', 3),
    ('Medici', 'Albizzi', 1),
    ('Medici', 'Salviati', 2),
    ('Salviati', 'Pazzi', 1),
    ('Peruzzi', 'Strozzi', 2),
    ('Ridolfi', 'Strozzi', 1),
    ('Ridolfi', 'Tornabuoni', 1),
    ('Strozzi', 'Bischieri', 1),
    ('Strozzi', 'Guadagni', 2),
    ('Bischieri', 'Guadagni', 1),
    ('Guadagni', 'Tornabuoni', 1)
    # Thm cc cnh khc vi trng s nu cn da trn cu trc  th gc
    # m bo bao gm tt c cc cnh trong  th gc vi trng s c gn
]

# Thm cc cnh c trng s vo  th mi
florentine_weighted.add_weighted_edges_from(weighted_edges)

# Bn c th cn m bo tt c cc nt v cnh gc u c bao gm,
# ngay c khi trng s ca chng ngm nh l 1 trong  th gc.
# i vi cc cnh khng c trong weighted_edges mt cch r rng, bn c th thm chng vi trng s mc nh.
original_edges = list(florentine_undirected.edges())
for u, v in original_edges:
    if not florentine_weighted.has_edge(u, v):
        # Thm cc cnh b thiu vi trng s mc nh, v d: 1
        florentine_weighted.add_edge(u, v, weight=1)


# Trc quan ha  th c trng s vi trng s cnh
pos = nx.spring_layout(florentine_weighted, seed=42) # S dng mt layout
plt.figure(figsize=(10, 8))
nx.draw(florentine_weighted, pos, with_labels=True, node_color='gold', node_size=800)
labels = nx.get_edge_attributes(florentine_weighted, 'weight')
nx.draw_networkx_edge_labels(florentine_weighted, pos, edge_labels=labels)
plt.title(" th Gia nh Florentine C Trng s (Trng s V d)")
plt.show()


deg_cent_w = nx.degree_centrality(florentine_weighted) #  o trung tm bc trn  th c trng s xem xt trng s nu c ch nh trong hm
# i vi  o trung tm bc vi trng s, networkx c mt hm c th:
weighted_degree = dict(florentine_weighted.degree(weight='weight'))
print("Bc C Trng s:", weighted_degree)

#  o trung tm gn gi trn  th c trng s s dng trng s cnh cho  di ng i
close_cent_w = nx.closeness_centrality(florentine_weighted, distance='weight')
print("\n o Trung tm Gn gi (C Trng s):", close_cent_w)

#  o trung tm trung gian trn  th c trng s s dng trng s cnh cho ng i ngn nht
bet_cent_w = nx.betweenness_centrality(florentine_weighted, weight='weight')
print("\n o Trung tm Trung gian (C Trng s):", bet_cent_w)

#  o trung tm eigenvector (kim tra ti liu  bit cch din gii trng s)
eig_cent_w = nx.eigenvector_centrality(florentine_weighted, weight='weight')
print("\n o Trung tm Eigenvector (C Trng s):", eig_cent_w) # B bnh lun  chy
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# 1. Ly thng tin t  th Cu lc b Karate
karate = nx.karate_club_graph()
n_karate = karate.number_of_nodes()
m_karate = karate.number_of_edges()
print(f"Cu lc b Karate - S nt: {n_karate}, S cnh: {m_karate}")

# c tnh xc sut p cho  th ngu nhin da trn s lng cnh
# i vi mt  th ngu nhin c n nt, s cnh ti a c th l n*(n-1)/2
# Vy, m = p * n*(n-1)/2  =>  p = m / (n*(n-1)/2)
# Thm mt epsilon nh vo p  tng nh kh nng kt ni,
# hoc gi nguyn  so snh s lng cnh trc tip hn (c th dn n mt kt ni)
p_karate = m_karate / (n_karate * (n_karate - 1) / 2)
print(f"Xc sut c tnh cho  th ngu nhin tng ng: {p_karate:.4f}")


# 2. To mt  th ngu nhin (m hnh ErdsRnyi G(n,p))
# Chng ta s s dng xc sut p  c tnh
random_graph = nx.erdos_renyi_graph(n_karate, p_karate)
print(f" th Ngu nhin c to - S nt: {random_graph.number_of_nodes()}, S cnh: {random_graph.number_of_edges()}")
# Lu : S lng cnh trong  th ngu nhin s gn bng m_karate, nhng c th khng hon ton ging do tnh cht xc sut.


# 3. Tnh ton cc s liu cho c hai  th

# Phn phi bc
degrees_karate = [deg for node, deg in karate.degree()]
degrees_random = [deg for node, deg in random_graph.degree()]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(degrees_karate, bins=np.arange(0, max(max(degrees_karate), max(degrees_random))+2)-0.5, color='steelblue')
plt.title("Phn phi Bc (Cu lc b Karate)")
plt.xlabel("Bc")
plt.ylabel("Tn sut")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(degrees_random, bins=np.arange(0, max(max(degrees_karate), max(degrees_random))+2)-0.5, color='salmon')
plt.title("Phn phi Bc ( th Ngu nhin)")
plt.xlabel("Bc")
plt.ylabel("Tn sut")
plt.grid(True)

plt.tight_layout()
plt.show()

# H s cm trung bnh
avg_clustering_karate = nx.average_clustering(karate)
avg_clustering_random = nx.average_clustering(random_graph)

print(f"\nH s Cm Trung bnh - Cu lc b Karate: {avg_clustering_karate:.4f}")
print(f"H s Cm Trung bnh -  th Ngu nhin: {avg_clustering_random:.4f}")

#  di ng i ngn nht trung bnh (Yu cu  th phi c kt ni)
# SA LI: Kim tra tnh kt ni trc khi tnh ton
print("\n---  di ng i Ngn nht Trung bnh ---")
if nx.is_connected(karate):
    try:
        avg_path_karate = nx.average_shortest_path_length(karate)
        print(f" di ng i Ngn nht Trung bnh - Cu lc b Karate: {avg_path_karate:.4f}")
    except nx.NetworkXNoPath:
         print("Cu lc b Karate b mt kt ni, khng th tnh ton  di ng i ngn nht trung bnh.")
else:
    print("Cu lc b Karate b mt kt ni, khng th tnh ton  di ng i ngn nht trung bnh.")


if nx.is_connected(random_graph):
    try:
        avg_path_random = nx.average_shortest_path_length(random_graph)
        print(f" di ng i Ngn nht Trung bnh -  th Ngu nhin: {avg_path_random:.4f}")
    except nx.NetworkXNoPath:
        # iu ny khng nn xy ra nu is_connected l True, nhng c bao gm  m bo tnh mnh m
        print(" th Ngu nhin b mt kt ni, khng th tnh ton  di ng i ngn nht trung bnh.")
else:
    print(" th Ngu nhin b mt kt ni, khng th tnh ton  di ng i ngn nht trung bnh.")


#  o Trung tm (v d:  o Trung tm Trung gian ca mt vi nt)
# t quan trng vic so snh cc nt c th hn l so snh s phn phi ca cc gi tr
bet_cent_karate = nx.betweenness_centrality(karate)
bet_cent_random = nx.betweenness_centrality(random_graph)

# In cc gi tr ln nht  so snh n gin
print(f"\n o Trung tm Trung gian Ln nht - Cu lc b Karate: {max(bet_cent_karate.values()):.4f}")
print(f" o Trung tm Trung gian Ln nht -  th Ngu nhin: {max(bet_cent_random.values()):.4f}")

# Lu :  so snh s phn phi  o trung tm mt cch cht ch hn,
# bn c th v biu  tn sut ca cc gi tr  o trung tm cho c hai  th.

# 4. So snh (Cc quan st t kt qu u ra)
# - Phn phi Bc:  th ngu nhin c xu hng c phn phi bc hp xung quanh bc trung bnh.
#   Cc mng trong th gii thc nh Cu lc b Karate thng c phn phi lch hn (mt vi nt c bc rt cao).
# - H s Cm Trung bnh: Cc mng trong th gii thc thng c h s cm trung bnh cao hn nhiu
#   so vi  th ngu nhin c cng s lng nt v cnh. iu ny phn nh s hin din ca nhiu tam gic
#   (bn ca bn cng l bn) trong cc mng x hi.
# -  di ng i Ngn nht Trung bnh: C hai loi mng u c th th hin hin tng "th gii nh"
#   ( di ng i trung bnh tng i ngn). Tuy nhin, s tng tc vi h s cm l yu t chnh.
# -  o Trung tm: S phn phi v phm vi ca cc gi tr  o trung tm c th khc nhau,
#   phn nh cc thuc tnh cu trc khc nhau v s hin din ca cc nt c nh hng.
```
